```python
import requests
from bs4 import BeautifulSoup

# URL de la página
url = 'https://www.fega.gob.es/atom/2024/rec_2024/'

# Hacer la solicitud GET a la página
response = requests.get(url, verify=False)

# Inicializar una lista para almacenar los enlaces
lista_enlaces = []

# Verificar si la solicitud fue exitosa
if response.status_code == 200:
    # Analizar el contenido HTML de la página
    soup = BeautifulSoup(response.content, 'html.parser')
    
    # Encontrar todos los elementos <a> (enlaces)
    enlaces = soup.find_all('a')

    # Extraer los enlaces (atributo href) y agregarlos a la lista
    for enlace in enlaces:
        href = enlace.get('href')
        if href:  # Asegurarse de que el enlace no sea None
            lista_enlaces.append(href)
else:
    print(f"No se pudo acceder a la página. Código de estado: {response.status_code}")

lista_enlaces = lista_enlaces[4:]
# Mostrar la lista de enlaces guardada (opcional)
print(lista_enlaces[1])
```

```python
import requests
from bs4 import BeautifulSoup
import numpy as np

#provincias = [6, 10, 22, 44, 50, 8, 17, 25, 43]
provincias = [43,45]
url_recintos = []
for i in provincias:
    url = f'https://www.fega.gob.es/atom/2024/rec_2024/{lista_enlaces[i]}'
    print(url)
  
    # Realiza una solicitud GET desactivando la verificación de SSL
    response = requests.get(url,verify=False)  # Cuidado con el riesgo de seguridad
    response.raise_for_status()

    # Utiliza BeautifulSoup con el analizador html.parser
    soup = BeautifulSoup(response.content, 'html.parser')

    # Encuentra todos los elementos 'link' y extrae el atributo 'href'
    links = [link.get('href') for link in soup.find_all('link')]


    # Encontrar todos los enlaces que contienen 'gpkg.zip'
    links = soup.find_all('a', href=True)
    enlaces = [link['href'] for link in links if '_gpkg.zip' in link['href']]
    
    # Mostrar los enlaces encontrados
    for gpkg_link in enlaces:
        print(gpkg_link)
        url_recintos.append(url+gpkg_link)
```

```python
import zipfile
import os
from io import BytesIO


# Directorio donde se guardarán los archivos descomprimidos
directorio_destino = f'/home/christianr/Documents/SIGPAC/recintos_SIGPAC/43' #office

# Crea el directorio destino si no existe
if not os.path.exists(directorio_destino):
    os.makedirs(directorio_destino)

for enlace in url_recintos:
    # Realiza la petición para descargar el archivo
    respuesta = requests.get(enlace, verify=False)
    # Abre el archivo ZIP descargado en memoria
    with zipfile.ZipFile(BytesIO(respuesta.content)) as archivo_zip:
        # Extrae todos los archivos en el directorio destino
        archivo_zip.extractall(directorio_destino)

print("Descarga y descompresión completadas.")
```
